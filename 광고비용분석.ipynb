```python
# 노출
imp = 10000
# 클릭
clk = 100
# 구매
conv = 10
# 비용
cost = 100000
```

```python
# ctr(click through rate) = 클릭수/노출수 *100
ctr = clk/imp*100
```

```python
ctr
```

```python
# cpm(click per mile) = 광고비용/노출수 *1000
cpm = cost/imp *1000
```

```python
cpm
```

```python
# cpc (cost per click) = 광고비용/클릭수
cpc = cost/clk
```

```python
cpc
# 클릭당 광고비용은 1000원 정도로 나왔다.
```

```python
# cpa (cost per action) = 광고비용 /구매수
cpa = cost /conv
```

```python
cpa
```

### 중점 관리 키워드 추출

    - 1. 95백분위수 찾기(quantile함수 사용)
    - 2. 95백분위수 이상의 노출수 추출
    - 3. 상위 5%에 해당하는 키워드 추출

```python
imp =df['노출수']
# quantile() => defalt는 median 중앙값
imp.quantile()
```

```python
#quantile(0) 최소값
imp.quantile(0)
```

```python
# quantile(1) 최댓값
imp.quantile(1)
```

```python
imp.quantile(0.95)
```

```python
# 상위 5%의 노출수만 추출
imp = imp[imp>=imp.quantile(0.95)]
```

```python
# 실습예제 원본
df = pd.read_excel('./네이버보고서.xls',skiprows=[0])
df.head()
```

```python
df_index = df.set_index('키워드')
```

```python
imp = df_index['노출수']
```

```python
imp = imp[imp>=imp.quantile(0.95)]
```

```python
imp
```

```python
imp = df_index['노출수']
clk = df_index['클릭수']
```

```python
result = df_index[(imp>=imp.quantile(0.95))&(clk>=clk.quantile(0.95))]
```

```python
result
```

- 노출수, 클릭수 기준 상위 5%의 키워드가 아닌데도 불구하고 비용이 많이 쓰이고 있는 키워드

```python
cost = df_index['총비용(VAT포함,원)']
```

```python
# 조건 4개 걸기
result = df_index[(imp<imp.quantile(0.95))&(clk<clk.quantile(0.95))&(cost>=cost.quantile(0.85))&(cost<cost.quantile(0.95))]
```

```python
result.index
```
